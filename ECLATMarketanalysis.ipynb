```python
import pandas as pd
from itertools import combinations

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/MLT/Market_Basket_Optimisation.csv')

# Convert rows into transactions
transactions = []

for _, row in df.iterrows():
    transaction = [str(item).strip() for item in row if pd.notna(item)]
    transactions.append(transaction)

# Create TID sets
item_tidset = {}

for tid, transaction in enumerate(transactions):
    for item in transaction:
        if item not in item_tidset:
            item_tidset[item] = set()
        item_tidset[item].add(tid)

# Minimum support
min_support = 3

# Frequent 1-itemsets
frequent_itemsets = {}

for item, tids in item_tidset.items():
    if len(tids) >= min_support:
        frequent_itemsets[(item,)] = len(tids)

# Frequent 2-itemsets
items = list(item_tidset.keys())

for item1, item2 in combinations(items, 2):
    common_tids = item_tidset[item1] & item_tidset[item2]

    if len(common_tids) >= min_support:
        frequent_itemsets[(item1, item2)] = len(common_tids)

# Display results
for itemset, support in frequent_itemsets.items():
    print(itemset, "Support =", support)
```
