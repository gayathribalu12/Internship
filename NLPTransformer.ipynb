```python
from transformers import pipeline
Classifier=pipeline("sentiment-analysis")
Classifier("This is the worst day in my life")
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "I am working on new school",
    candidate_labels=["education","politics","business"]
)
```

```python
generator=pipeline("text-generation")
generator("In this course, we will teach you how to")
```

```python
generator=pipeline("text-generation",model="openbmb/MiniCPM5-1B")
generator(
    "In this class,i will teach you to",
    max_lenghth=40,
    num_return_sequences=4,
)
```

```python
generator=pipeline("fill-mask",model="google-bert/bert-base-uncased")
generator(
    "Machine learning is a branch of <mask> intelligence."
)
```
