```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth

# Read dataset
df = pd.read_csv('/content/drive/MyDrive/MLT/Market_Basket_Optimisation.csv')

# Convert rows into transactions
transactions = []

for _, row in df.iterrows():
    transaction = [str(item) for item in row.dropna()]
    transactions.append(transaction)

# One-Hot Encoding
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df_encoded = pd.DataFrame(te_array, columns=te.columns_)

# FP-Growth
frequent_itemsets = fpgrowth(df_encoded,
                             min_support=0.03,
                             use_colnames=True)

print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets,
                          metric='confidence',
                          min_threshold=0.5)

print(rules[['antecedents',
             'consequents',
             'support',
             'confidence',
             'lift']])
```
